```python
class Equipo:

    def __init__(self, nombre, camiseta, estadio, provincia, apodo, dir_tecnico, titulos):

        self.nombre_equipo = nombre
        self.color_camiseta = camiseta
        self.estadio = estadio
        self.provincia = provincia
        self.apodo = apodo
        self.dir_tecnico = dir_tecnico
        self.titulos = titulos
        self.jugadores = []


    def info_equipo(self):

        print(f" Equipo:\n {self.nombre_equipo}")
        print(f".Color de camiseta: {self.color_camiseta}")
        print(f".Estadio: {self.estadio}")
        print(f".Provincia: {self.provincia}")
        print(f".Apodo: {self.apodo}")
        print(f".Dir. Tecnico: {self.dir_tecnico}")
        print(f".Titulos: {self.titulos}")


    def agregar_jugador(self, jugador):
        self.jugadores.append(jugador)
        print(f"El jugador {jugador.nombre} se ha unido a {self.nombre_equipo}")



    def mostrar_jugador(self):
        print(f"\nPlantel de {self.nombre_equipo}:")
        for jugador in self.jugadores:
            print(f"- {jugador.nombre} ({jugador.posicion})")

    def sumar_titulo(self):

        self.titulos += 1
        print(f"{self.nombre_equipo} ha ganado un nuevo titulo")


    def cambiar_dt(self, nuevo_dt):
        self.dir_tecnico = nuevo_dt
        print(f"{nuevo_dt} es el nuevo tecnico de {self.nombre_equipo}")



Belgrano = Equipo("Club Atletico Belgrano" , "Celeste", "Julio Cesar Villagra" , "Cordoba", "Los piratas" , "Ruso Zielinski", 1)
Talleres = Equipo("Club Atletico Talleres" , "Azul y blanco" , "La Boutique" , "Cordoba", "El matador" , "Carlos Tevez", 2)
River = Equipo("Club Atletico River Plate" , "Rojo y blanco", "Mas monumental" , "Buenos Aires", "El millonario" , "Marcelo Gallardo", 72 )
Boca = Equipo("Club Atletico Boca Juniors" , "Azul y oro", "La bombonera" , "Buenos Aires", "Xeneize" , "Miguel Angel Russo", 74)






class Jugador:

    def __init__(self, nombre, posicion, ataque, defensa , estado_fisico):

        self.nombre = nombre
        self.posicion = posicion
        self.ataque = ataque
        self.defensa = defensa
        self.estado_fisico = estado_fisico
        self.lesionado = False


    def info_jugador(self):

        print(f" Nombre:\n {self.nombre}")
        print(f".Posicion: {self.posicion}")
        print(f".Ataque: {self.ataque}")
        print(f".Defensa: {self.defensa}")
        print(f".Estado Fisico: {self.estado_fisico}")
        print(f".Lesionado: {'Sí' if self.lesionado else 'No'}")


    def entrenar(self):
        self.estado_fisico = max(0, self.estado_fisico - 2)
        print(f"{self.nombre} ha completado su entrenamiento, disminuyendo 2 puntos su estado fisico.")


    def lesion(self):
        self.estado_fisico = max(0, self.estado_fisico - 30)
        if self.estado_fisico > 60:
            print(f"El jugador {self.nombre} esta siendo asistido por medicos")
            print(f"{self.nombre} ingreso nuevamente al partido")

        elif self.estado_fisico > 30 and self.estado_fisico <= 60:
            print(f"{self.nombre} salio lesionado y se perdera una fecha")
            self.lesionado = True

        else:
            print(f"{self.nombre} sufrio una lesion grave y se perdera varias fechas")
            self.lesionado = True



    def recuperarse(self):
        self.estado_fisico = min(100, self.estado_fisico + 20)

        if self.estado_fisico > 60:
            self.lesionado = False
            print(f"El jugador {self.nombre} volvió a entrenar con el plantel")

        elif self.estado_fisico > 30:
            self.lesionado = False
            print(f"{self.nombre} se está entrenando de manera diferenciada")

        else:
            print(f"{self.nombre} sigue lesionado y necesita más tiempo de recuperación")


class Delantero(Jugador):

    def __init__(self, nombre, posicion, ataque, defensa, estado_fisico, goles= 0):
        super().__init__(nombre, posicion, ataque, defensa, estado_fisico)
        self.goles = goles



    def info_jugador(self):
        super().info_jugador()
        print(f".Goles: {self.goles}")

    def entrenar(self):
        super().entrenar()
        self.ataque += 2
        print(f"{self.nombre} ha aumentado +2 su ataque")


    def lesion(self):
        super().lesion()
        self.ataque -= 15
        print(f"{self.nombre} ha perdido 15 puntos en ataque")


    def anotar_gol(self, cantidad_goles):
        self.goles += cantidad_goles

        if cantidad_goles == 1:
            print(f"{self.nombre} ha anotado {cantidad_goles} gol")
        elif cantidad_goles == 2:
            print(f"{self.nombre} ha anotado {cantidad_goles} goles")
        elif cantidad_goles == 3:
            print(f"{self.nombre} ha anotado un HAT TRICK!!!!")



class Mediocampista(Jugador):

    def __init__(self, nombre, posicion, ataque, defensa, estado_fisico, asistencia = 0):
        super().__init__(nombre, posicion, ataque, defensa, estado_fisico)
        self.asistencia = asistencia

    def info_jugador(self):
        super().info_jugador()
        print(f".Asistencias: {self.asistencia}")


    def entrenar(self):
        super().entrenar()
        self.ataque += 1
        self.defensa += 1
        print(f"{self.nombre} ha aumentado +1 su ataque y +1 su defensa")



    def lesion(self):
        super().lesion()
        self.defensa -= 8
        self.ataque -= 8
        print(f"{self.nombre} ha perdido 8 puntos en ataque y defensa")


    def asistir(self):
        self.asistencia += 1
        print(f"{self.nombre} ha dado una asistencia")



class Defensor(Jugador):

    def __init__(self, nombre, posicion, ataque, defensa, estado_fisico, expulsion= 0):
        super().__init__(nombre, posicion, ataque, defensa, estado_fisico)
        self.expulsion = expulsion


    def info_jugador(self):
        super().info_jugador()
        print(f".Expulsiones: {self.expulsion}")

    def entrenar(self):
        super().entrenar()
        self.defensa +=2
        print(f"{self.nombre} ha aumentado +1 su defensa")


    def lesion(self):
         super().lesion()
         self.defensa -= 15
         print(f"{self.nombre} ha perdido 15 puntos en defensa")

    def recibir_expulsion(self):
        self.expulsion += 1
        print(f"{self.nombre} ha sido expulsado")

        if self.expulsion == 2:
            print(f"{self.nombre} debe cumplir 2 fechas de suspension")




#Plantel de Belgrano
Jara = Delantero("Franco Jara", "Delantero", 80, 30, 71)
Zela = Mediocampista("Lucas Zelarrayan", "Mediocampista", 70, 45, 75)
Romero = Defensor("Cristian Romero", "Defensor", 30, 80, 78)

#Plantel de Talleres
Girotti  = Delantero("Federico Girotti", "Delantero", 76, 37, 82)
Botta = Mediocampista("Ruben Botta", "Mediocampista", 68, 42, 79)
Catalan = Defensor("Matias Catalan" , "Defensor", 27, 69, 73)


#Plantel de River
Salas = Delantero("Maxi Salas", "Delantero", 81, 35, 84)
Quintero = Mediocampista("Juan Fernando Quintero", "Mediocampista", 71, 53, 68)
Diaz = Defensor("Paulo Diaz" , "Defensor", 32, 73, 64)

#Plantel de Boca
Merentiel = Delantero("Miguel Merentiel", "Delantero", 80, 35, 82)
Paredes = Mediocampista("Leandro Paredes", "Mediocampista", 71, 53, 84)
Blanco = Defensor("Lautaro Blanco" , "Defensor", 32, 73, 64)


#Agregando jugadores a Belgrano
Belgrano.agregar_jugador(Jara)
Belgrano.agregar_jugador(Zela)
Belgrano.agregar_jugador(Romero)

#Agregando jugadores a Talleres
Talleres.agregar_jugador(Girotti)
Talleres.agregar_jugador(Botta)
Talleres.agregar_jugador(Catalan)

#Agregando jugadores a River
River.agregar_jugador(Salas)
River.agregar_jugador(Quintero)
River.agregar_jugador(Diaz)


#Agregando jugadores a Boca
Boca.agregar_jugador(Merentiel)
Boca.agregar_jugador(Paredes)
Boca.agregar_jugador(Blanco)

Belgrano.mostrar_jugador()

print("-----------------")

Talleres.mostrar_jugador()

print("-----------------")

River.mostrar_jugador()

print("-----------------")

Boca.mostrar_jugador()


Jara.info_jugador()
Jara.lesion()
Jara.info_jugador()

Blanco.lesion()








```

```python
import random
import time

def simular_partido(equipo1, equipo2):
    print(f"\nPartido: {equipo1.nombre_equipo} vs {equipo2.nombre_equipo}\n")
    time.sleep(1)

    # Calcular ataque y defensa de cada equipo
    ataque1 = sum(j.ataque for j in equipo1.jugadores if j.posicion in ["Delantero", "Mediocampista"])
    defensa1 = sum(j.defensa for j in equipo1.jugadores if j.posicion in ["Defensor", "Mediocampista"])
    ataque2 = sum(j.ataque for j in equipo2.jugadores if j.posicion in ["Delantero", "Mediocampista"])
    defensa2 = sum(j.defensa for j in equipo2.jugadores if j.posicion in ["Defensor", "Mediocampista"])

    # Determinar goles con factor aleatorio
    goles1 = max(0, (ataque1 - defensa2) // 20 + random.randint(0, 2))
    goles2 = max(0, (ataque2 - defensa1) // 20 + random.randint(0, 2))

    # Listas de delanteros
    delanteros1 = [j for j in equipo1.jugadores if j.posicion == "Delantero"]
    delanteros2 = [j for j in equipo2.jugadores if j.posicion == "Delantero"]

    # Simulación de acciones del equipo 1
    for i in range(goles1):
        time.sleep(1)
        if delanteros1:
            goleador = random.choice(delanteros1)
            goleador.anotar_gol(1)
            print(f"¡{goleador.nombre} anotó un gol para {equipo1.nombre_equipo}!")

    # Simulación de acciones del equipo 2
    for i in range(goles2):
        time.sleep(1)
        if delanteros2:
            goleador = random.choice(delanteros2)
            goleador.anotar_gol(1)
            print(f"¡{goleador.nombre} anotó un gol para {equipo2.nombre_equipo}!")

    print(f"\nResultado final: {equipo1.nombre_equipo} {goles1} - {goles2} {equipo2.nombre_equipo}\n")
    time.sleep(1)

    # Reducir estado físico y aplicar posibles lesiones
    for jugador in equipo1.jugadores + equipo2.jugadores:
        decremento = random.randint(3,7)
        jugador.estado_fisico = max(0, jugador.estado_fisico - decremento)
        if jugador.estado_fisico <= 60:
            time.sleep(1)
            jugador.lesion()

    # Mostrar estado final de los jugadores
    print("\nEstado final de los jugadores:")
    for jugador in equipo1.jugadores + equipo2.jugadores:
        time.sleep(0.5)
        jugador.info_jugador()
        print()
```

```python
simular_partido(Belgrano, Talleres)
```

```python
#Grafico de tabla de goleadores

import matplotlib.pyplot as plt

# Lista de delanteros
delanteros = [Jara, Girotti, Salas, Merentiel]

# Extraer nombres y goles
nombres = [jugador.nombre for jugador in delanteros]
goles = [jugador.goles for jugador in delanteros]
```

```python
plt.figure(figsize=(8,5))
plt.barh(nombres, goles, color='skyblue', edgecolor='black')

plt.title("Tabla de Goleadores", fontsize=14, color='navy')
plt.xlabel("Cantidad de Goles", fontsize=12)
plt.ylabel("Jugador", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.gca().invert_yaxis()  # Para que el que más goles tiene aparezca arriba

plt.show()
```
